```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import lxml
import re
```

```python
url_bundles = "https://stardewvalleywiki.com/Crops"
response = requests.get(url_bundles)
content = BeautifulSoup(response.content, "html.parser")
findable = content.find_all('table')


# Grab the html code that houses the buy points

buy_html = []

for table in findable:
    rows = table.find_all('tr')

    for row in rows:
        cells = row.find_all('td')

        for search in cells:

            still_looking = search.find_all('div')

            for i in still_looking:
                sell_points = i.find_all('span',{'class': 'no-wrap'})
                buy_html.append(sell_points)

# Removes all blank entries                
res = [ele for ele in buy_html if ele != []]

try1 = []
buy_prices = []

# Puts each line as a string
for list in res:
    for i in list:
        try1.append(str(i))

# Removes html fluff to get the actual gold price
for i in try1:

    result = re.search('data-sort-value="(.*)"><img alt=', i)

    buy_prices.append(result.group(1))

# Grab every other value, we don't want Joja Corp's prices >:|
#buy_prices = buy_prices[::2]
```
